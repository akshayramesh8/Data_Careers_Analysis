<font size= 7>1. How many jobs are currently available in the Indian market?</font>


```python
# 1) How many tech jobs are available in the Indian market?

df_india['job_title_short'].size
```

<font size=5>There are 51088 job opportunities available in the Indian job market.</font>

```python
# 2) How many jobs are posted for each unique role?

ax = df_india['job_title_short'].value_counts(ascending= True).plot(kind='barh')
plt.ylabel('')
ax.bar_label(ax.containers[0], padding= 3, fmt= '%d', fontweight= 'medium')
plt.title('Tech Job Postings in India', fontweight= 'semibold' , fontsize= 15)

for spine in ['top' , 'right']:
  ax.spines[spine].set_visible(False)

plt.show()
```

<font size= 4> It is interesting to note that Data roles are more in demand than conventional Software Developer roles!</font>

```python
# Let's explore the top data roles in the country

job_count = df_india['job_title_short'].value_counts(ascending= False).head(5)

job_titles = job_count.index
job_values = job_count.values

plt.pie(
    job_values,
    labels=job_titles,
    autopct='%1.1f%%',
    pctdistance=0.85,
    wedgeprops={'edgecolor': 'black'},
    # labeldistance= 1.2,
    textprops={'fontsize': 10, 'fontweight': 'light'},
    startangle=140
)

plt.gca().set_aspect('equal')
plt.title("Top 5 Data Job Roles in India", fontweight='bold')

plt.show()

# Add leader lines (ticks)
# plt.legend(title="Job Roles", loc="best", bbox_to_anchor=(1, 1))  # Optional legend
```

```python
# Let's explore the top data roles in the country

job_count = df_india['job_title_short'].value_counts(ascending= False).head(5)

job_titles = job_count.index
job_values = job_count.values

plt.pie(
    job_values,
    labels=job_titles,
    autopct='%1.1f%%',
    pctdistance=0.85,
    wedgeprops={'edgecolor': 'black'},
    # labeldistance= 1.2,
    textprops={'fontsize': 10, 'fontweight': 'light'},
    startangle=140
)

plt.gca().set_aspect('equal')
plt.title("Top 5 Data Job Roles in India", fontweight='bold')

plt.show()

# Add leader lines (ticks)
# plt.legend(title="Job Roles", loc="best", bbox_to_anchor=(1, 1))  # Optional legend
```

<font size= 7> 2. Which are the top jobs available in the data domain?</font>

```python
# 3) How many jobs are available in the data domain?

# df_india.groupby(df_india[df_india['job_title_short'] == '%Data%']).size() -- this will not work

df_india_data = df_india[df_india['job_title_short'].str.contains('Data', case= False, na= False)]

df_india_data.sample(5)
```

```python
data_jobs = df_india_data.groupby('job_title_short').size().sort_values(ascending= False)

ax = data_jobs.plot(kind='bar', color='darkcyan', edgecolor='black', linewidth=1.5)

# for p in ax.patches:
#     ax.annotate(f'{p.get_height()}',  # Get the height (value) of each bar
#                 (p.get_x() + p.get_width() / 2, p.get_height()),  # Position: center of the bar
#                 ha='center', va='bottom',  # Align text
#                 fontsize=10, color='black')

ax.bar_label(ax.containers[0], fmt='%d', padding=3, fontsize=10, fontweight='bold')

for spine in ['top', 'right']:
  ax.spines[spine].set_visible(False)


plt.title('Count of Data Jobs in India', fontsize= 14, fontweight= 'bold')
plt.xlabel('')
plt.xticks(rotation= 45, ha= 'right')
plt.show()


# plt.grid(axis='y', linestyle='--', alpha=0.7) for gridlines
```
